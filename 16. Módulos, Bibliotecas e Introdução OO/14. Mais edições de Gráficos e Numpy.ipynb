# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)  # gerar 50 números inteiros aleatórios, de 1000 a 2999 (o último é exclusivo)
meses = np.arange(1, 51)  # range sequencial de 1 a 50
print(meses)
```

```python
plt.plot(meses, vendas)
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
# mudando a linha para apenas marcadores
plt.plot(meses, vendas, '-')
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
plt.plot(meses, vendas, ':', c='orange')
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas, c='red')
plt.show()
```

- Barras

```python
plt.bar(meses, vendas, color='purple')
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize=(18, 3))
plt.subplot(1, 3, 1)  # sintaxe: subplot(nrows, ncols, index, **kwargs)
plt.plot(meses, vendas, '-')
plt.subplot(1, 3, 2) 
plt.scatter(meses, vendas, c='red')
plt.subplot(1, 3, 3)
plt.bar(meses, vendas, color='purple')
plt.show()
```
