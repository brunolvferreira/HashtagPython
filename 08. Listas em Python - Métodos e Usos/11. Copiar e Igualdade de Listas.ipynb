# Copiar e "Igualdade" de Listas

### Estrutura:

- Quando fazemos:<br>
lista2 = lista1<br>
não estamos criando uma lista nova, mas estamos atribuindo outra variável à mesma lista.

- Se quisermos copiar lista devemos fazer<br>
lista2 = lista1.copy()<br>
ou entao<br>
lista2 = lista1[:]

Para entender bem isso, vamos ver na prática:

```python
lista = ['ipad', 'iphone x', 'apple tv']
lista2 = lista  # lista 2 aponta para lista
print(lista)

lista[1] = 'iphone 11'

print(lista)
print(lista2)
```

### Agora copiando:

```python
lista = ['ipad', 'iphone x', 'apple tv']
lista2 = lista.copy()  # lista 2 recebe uma cópia da lista, se tornando uma variável independente
# lista2 = lista[:]  # outra opção de sintaxe

lista[1] = 'iphone 11'

print(lista)
print(lista2)
```
