# Juntar Listas, Ordenar e Cuidados Especiais

### 2 formas:

- lista1.extend(lista2)  (Edita a lista original)
- lista_nova = lista1 + lista2

Obs: o Método .append não junta listas, mas adiciona um valor no final da lista

```python
produtos = ['apple tv', 'mac', 'iphone x', 'iphone 11', 'IPad', 'apple watch', 'mac book', 'airpods']
novos_produtos = ['iphone 12', 'ioculos']

todos_produtos = produtos + novos_produtos
print(todos_produtos)  # listas não foram alterada
print(produtos)
print(novos_produtos)

produtos.append(novos_produtos)  # adicionando um novo valor ao final da lista, a lista novos_produtos
print(produtos)  # retorna 1 lista com outra lista dentro, como um único valor

produtos.extend(novos_produtos)
print(produtos)  # lista foi alterada
```

### Cuidado:

- [1] + [2] não é a mesma coisa que 1 + 2, então cuidado sempre com o formato dos valores na hora de fazer operações.

```python
vendas = [1000, 1500, 15000, 20000, 270, 900, 100, 1200]
vendas_iphonex = [15000]
vendas_iphone11 = [20000]

print([1] + [2])  # gera uma lista [1, 2]
print(1 + 2)

total_iphone = vendas[2] + vendas[3]
total_iphone_lista = vendas_iphonex + vendas_iphone11
total_iphone_lista2 = vendas_iphonex[0] + vendas_iphone11[0]

print(total_iphone)  # 35000
print(total_iphone_lista)  # [15000, 20000]
print(total_iphone_lista2)  # 35000
```

### Ordenar listas

lista.sort()

```python
produtos.sort()  # ordem alfabética, maiúsculas são distintas de minúsculas (odem da Tabela ASCII)
print(produtos)

vendas.sort()  # ordem do menor parao maior valor
print(vendas)

vendas.sort(reverse=True)  # true com T, ordendando do maior para o menor
# cuidado com sort(reverse=False)!!!
print(vendas)
```
