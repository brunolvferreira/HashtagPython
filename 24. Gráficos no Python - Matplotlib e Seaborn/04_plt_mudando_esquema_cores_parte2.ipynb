## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')  # bottom (empilhando os dados de vendas_online abaixo de vendas_loja)
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()  # rótulos (legenda), depende das labels para funcionar, necessário declarar label dentro do plt.bar
plt.show()  
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

plt.xticks(ticks=anos, labels=anos)  # forçar os anos a serem exibidos como inteiros
# plt.xticks(ticks=anos, labels=['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4'])  # opção para rotular o nome nos ticks

# plt.xlabel('Ano')  # neste caso é "dispensável"
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

cores = plt.get_cmap('Pastel2').colors  # escolhendo a paleta de cores e atribuindo a uma variável

plt.bar(anos, vendas_online, label='Online', color=cores[2])
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja', color=cores[3])

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos, labels=anos)

# plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler  # utilizada para aplicar o ciclo nas cores, caso haja mais dados que a quantidade de cores na paleta

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

# rc = runtime configuration (configuração quando executa)
cores_cycler = cycler('color', plt.get_cmap('Pastel1').colors)
plt.rc('axes', prop_cycle=cores_cycler)  # alterou toda paleta de cores do plt (até que seja redefinido mais abaixo)
 
plt.bar(anos, vendas_online, label='Online')
plt.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

# forçar os anos a serem exibidos como inteiros
plt.xticks(ticks=anos)

plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.title('Vendas Online vs Loja')
plt.legend()
plt.show()
```
