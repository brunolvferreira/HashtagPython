## 5.1 Ordem de Execução dos Programas
O código SEMPRE é executado de cima para baixo.

```python
faturamento = 1000  # precisa ser definido antes do print senão gera erro
custo = 500
lucro = faturamento - custo  # calculo1 de lucro (500)
print("Lucro 1 = ", lucro)  # retorna 500, custo estava em 500

custo = 700
# lucro = faturamento - custo  # calculo2 de lucro (700)
print("Lucro 2 = ", lucro)  # retorna 300, custo subiu para 700
# retorna 500 caso a linha acima seja comentada porque pega referência de cálculo do primeiro 'faturamento - custo'
```

## 5.2 Tipos de Variáveis
Uma variável é um objeto, afinal tudo é um objeto em Python.

int --> inteiro<br>
string --> texto<br>
float --> números com casas decimais (ponto flutuante)<br>
bool ou boolean --> True ou False<br>

Obs.: variáveis em Python não são declaradas explicitamente, mas isso não significa que você deve ficar mudando o tempo todo.<br>
Obs2.: cuidado com os nomes restritos

```python
faturamento = 1000
type(faturamento)  # int
```

```python
faturamento = 1000.00
type(faturamento)  # float
```

```python
faturamento = '1.000'
type(faturamento)  # str
```

```python
ganha_bonus = True
type(ganha_bonus)  # bool
```

## 5.3 Misturando Tipos de Variáveis

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo
# print('O faturamento da loja foi' + faturamento)  # TypeError: can only concatenate str (not "int") to str
print('O faturamento da loja foi ' + str(faturamento))  # transformando int em str
print('O custo da loja foi ' + str(custo))
print('Assim, o lucro da loja foi de ' + str(lucro) + ' reais')
```

## 5.4 Comando Format

```python
faturamento = 2000
custo = 500
lucro = faturamento - custo

print('O faturamento da loja foi {}. O custo foi {}. Assim, o custo ficou em {}.'.format(faturamento, custo, lucro))

# {1} {2} {3}.format(1, 2, 3)
```

## 5.5 F-string e Mudança de Tipo de Variável

```python
# Format vs. F-string

faturamento = 1000
custo = 400
lucro = faturamento - custo

# com format
print('O faturamento foi de {} e o lucro de {}'.format(faturamento, lucro))

#com f-string
print(f'O faturamento foi de {faturamento} e o lucro de {lucro}')
```

```python
# Mudança de Tipo de Variável
faturamento = float(input('Insira o faturamento '))  # input recebe um texto, faturamento seria reconhecido como str
custo = float(input('Insira o custo '))  # input recebe um texto, custo seria reconhecido como str
# declarando int antes do input é realizado a conversão para int, e assim por diante
# str ==> muda para string
# int ==> muda para inteiro
# float ==> muda para float (decimal sempre separado por ponto)

print(type(faturamento))
print(type(custo))

lucro = faturamento - custo
print(lucro)
```
